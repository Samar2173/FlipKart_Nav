```python
import pandas as pd
```

```python
filename = r'Files\Co-ordinates.csv'
df = pd.read_csv(r'E:\Projects\Flipkart E-Grid\Files\Co-ordinates.csv')
df.head(20)
```

```python
for i in range(1, len(df.x)):
    if abs(df.x[i] - df.x[i-1]) < 1.5:
        df.x[i] = df.x[i-1]
df.head(20)
```

```python
for i in range(1, len(df.y)):
    if abs(df.y[i] - df.y[i-1]) < 1.5:
        df.y[i] = df.y[i-1]
```

```python
df['x']
```

```python
df = df.drop_duplicates()
```

```python
df.sort_values(by = 'x')
```

```python
x_val = df.groupby('x')
print(x_val.size())
```

```python
x_val = df.groupby('y')
print(x_val.size())
```
